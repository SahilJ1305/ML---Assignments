```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
data = pd.read_csv("/content/temperatures.csv")
df = data
```

```python
x = data["YEAR"].values[:,None]
y = data["MAY"]

print(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'R-Square Score: {r2:.2f}')
```

```python
plt.scatter(X_train, y_train, color='red', label='Actual')
plt.plot(X_test, y_pred, color = "green", label = "Predicted")
plt.xlabel("May temperature")
plt.ylabel("Annual Temprature")
plt.show()
```
